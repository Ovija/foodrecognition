```python
from ultralytics import YOLO
import torch
import os
import wandb

# Set up the environment for Weights and Biases
os.environ['WANDB_API_KEY'] = 'xxx'
wandb.login()  # This will use the API key from the environment variable

# Load the model from the last checkpoint (use the path to the checkpoint file)
# model = YOLO('/kaggle/input/foodseg103-yolo-result/runs/detect/train/weights/last.pt')
model = YOLO('yolov8m-seg.pt')

# Train the model for an additional 20 epochs, making it 40 epochs in total
model.conf = 0.1

model.train(
    data='/kaggle/input/datasett/dataset_.yaml',  # Path to the dataset.yaml file
    epochs=40,  # Total number of epochs (since 20 were done previously, this adds 20 more)
    batch=4,  # Adjust based on GPU memory
    device=0 if torch.cuda.is_available() else 'cpu',
    patience=10,
    lr0=1e-3
)
```
