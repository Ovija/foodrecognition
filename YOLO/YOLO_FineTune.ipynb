```python
# import necessary packages
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import Dataset
import torch
import albumentations as A
from ultralytics import YOLO
from kaggle_secrets import UserSecretsClient
from ultralytics import YOLO
import os
import wandb
```

# YOLOv8m-Seg Training

```python
# Set up the environment for Weights and Biases
user_secrets = UserSecretsClient()
os.environ['WANDB_API_KEY'] = user_secrets.get_secret("WANDB_API_KEY")

wandb.login()  # This will use the API key from the environment variable

# Load the pretrained model for segmentation
# model = YOLO('yolov8m-seg.pt')

# Load the best model from last run for further fine-tuning
model = YOLO('/kaggle/input/foodseg103-yolov8m-seg-output/runs/segment/train/weights/best.pt')

# Adjust the Confidence threshold - set it lower than default
model.conf = 0.1

# Train/Fine-Tune the model
model.train(
    data='/kaggle/input/foodseg103-yolo-yaml/dataset_.yaml',  # Path to the dataset.yaml file
    epochs=40,  
    batch=4,  
    device=0 if torch.cuda.is_available() else 'cpu',
    patience=10, # early stopping after 10 epochs in order to avoid overfitting
    lr0=1e-5, # Use a smaller learning rate for fine-tuning
    cos_lr=True,   # Use a cosine annealing scheduler
    warmup_epochs=3,  # Number of epochs for warmup
    augment = True # Enable augmentation
)

# End the W&B logging session
wandb.finish()
```
